```python
import os
import ipywidgets as widgets
from ipywidgets import HBox, VBox
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
%matplotlib inline
```

```python
# Path for exported data, numpy arrays
DATA_PATH = os.path.join('data') 

# Actions that we try to detect
actions_list = np.array(['hello', 'thanks', 'iloveyou'])
```

```python
@widgets.interact(
    category=['hello', 'iloveyou', 'thanks'],
    sequence=(0,29), 
    frame=(0, 29)
)
def f(category="hello",sequence=0, frame=0):
    res = np.load(os.path.join(DATA_PATH, category, str(sequence), "{}.npy".format(frame)))
    # Generate some example data
    data_pose = res[0:33*3].reshape(33,3)
    data_face = res[33*4:468*3+33*4].reshape(468,3)
    data_lh = res[33*4+468*3 : 33*4+468*3+21*3].reshape(21,3)
    data_rh = res[33*4+468*3+21*3 : ].reshape(21,3)
    # Create a figure and axis object
    fig, ax = plt.subplots()

    # Plot the data as a scatter plot
#     ax.scatter(data_pose[:, 0], data_pose[:, 1])
    ax.scatter(data_face[:, 0], data_face[:, 1])
    ax.scatter(data_lh[:, 0], data_lh[:, 1])
    ax.scatter(data_rh[:, 0], data_rh[:, 1])
    # Set labels and title
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_title('Plot of a 33x3 NumPy array')

    # Show the plot
    plt.show()
```
